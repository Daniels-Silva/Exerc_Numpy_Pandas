# Analise Outliers
Considerando que os Outliers identificados são referentes a quantidade de crédito, e considero um dado importante para a analise, a sugestão seria analisar estes dados separadamente, ou seja a criação de uma nova base, com os clientes que possuem crédito acima de XXXXX (Outliers), para isso iremos utilizar a função criada na aula 6

```
# Função para identificar Outliers aula 06

def identificar_outliers(df, coluna):
    # Calculando o primeiro e terceiro quartis
    Q1 = df[coluna].quantile(0.25) #Primeiro Quartil
    Q3 = df[coluna].quantile(0.75) #Terceiro Quartil
    
    # Calculando o IQR (Intervalo Interquartil)
    IQR = Q3 - Q1
    
    # Calculando os limites superior e inferior para identificar os outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Identificando outliers na coluna especificada
    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    
    return outliers
```

```
# Utilizando a função e criando uma nova base, para analise dos Outliers separadamente.

analiseOutliers = identificar_outliers(basesJuntas,'Credit amount')
analiseOutliers
```
